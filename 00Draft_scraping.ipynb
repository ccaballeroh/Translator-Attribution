# Draft

This is a draft notebook only ran once to scrap a list of cohesion markers from this [website](http://home.ku.edu.tr/~doregan/Writing/Cohesion.html). The list was later manually curated and modified and can be found in the `auxfiles/json` folder on this repository in the file `markersList.json`.

The purpose of having this notebook here is just for documentation sake.

```python
import requests, bs4
import json
```

```python
res = requests.get("http://home.ku.edu.tr/~doregan/Writing/Cohesion.html")
```

```python
res.raise_for_status()
```

```python
noStarchSoup = bs4.BeautifulSoup(res.text)
```

```python
markers = noStarchSoup.select('table')[8].getText()
```

```python
markers[:100]
```

```python
JSON_FOLDER = ".\\auxfiles\\json\\"
```

```python
with open(JSON_FOLDER + "markers_list.json", "w") as f:
    json_text = json.dumps(markers.lower().split())
    f.write(json_text)
    print("file saved!")
```

```python
with open(JSON_FOLDER + "markers_list.json", "r") as g:
    markers = json.loads(g.read())
```

```python
markers.sort()
```

```python
markers
```

```python
markers_redux = list(set(markers))
```

```python
markers_redux.sort()
```

```python
markers_redux
```

```python
with open(JSON_FOLDER + "markers_list_redux.json", "w") as f:
    json_text = json.dumps(markers_redux)
    f.write(json_text)
    print("file saved!")
```
